# Consistent, inconsistent and indeterminate systems

So far all the examples of systems of linear equations we have seen have had a single unique solution. This will not always be the case, and it is important that we consider the different situations that can arise when dealing with systems of equations.

For example, there might not be enough information given by the equations to be able to fully determine the solution. Or it may be that there are no possible solutions, as the equations contradict each other. While it is possible to just look at the equations to determine if a system is solvable or not, an easier way is to examine the corresponding matrix of coefficients and augmented matrix.

One measure that can be helpful for this is the **rank** of a matrix.

```{index} Matrix ; rank
```

```{index} Rank
```

```{prf:definition} Rank of a matrix
:label: rank-definition

The **rank** of a matrix $A$ denoted by $\operatorname{rank}(A)$ is the number of non-zero rows of $A$ when it has been converted to row echelon form. 
```

To compute the rank of a matrix, we can use elementary row operations to row reduce to row echelon form, and then count the number of non-zero rows. Comparing the rank to the overall number of rows in the matrix will help us determine if the system has solutions, as outlined below.

Since linear equations represent straight lines, we can visualise the solutions - and whether they will exist or not - by considering the intersections of the lines. We can categorise linear systems of equations in three ways:
```{prf:definition} Consistent, inconsistent and indeterminate systems
:label: consistent-inconsistent-and-indeterminate-systems-definition

A system of equations is said to be **consistent** if it has a solution, and otherwise it is said to be **inconsistent**. If a system of equations has more than one solution, it is said to be **indeterminate**.
```

```{index} Linear systems ; consistent system
```

## Consistent systems

```{index} Systems of linear equations ; consistent systems
```

Consider the plot of a system with two variables, $x_1$ and $x_2$. If the lines that represent each equation in a system intersect at a single point, then the coordinates of the point of intersection provide a unique solution to the system. Consider the following system of linear equations:

$$ \begin{align*}
    x_1 - x_2 &= 1, \\
    x_1 + x_2 &= 3.
\end{align*} $$

The lines for these two equations have been plotted in {numref}`consistent-system-plot` below.

```{glue:figure} consistent_system_plot
:figwidth: 500
:name: consistent-system-plot

Plots of the lines $x_1 + x_2 = 3$ and $x_1 - x_2 = 1$.
```

The two lines intersect at $(2,1)$ so this system is **consistent** and has a unique solution. This can easily be confirmed using Gaussian elimination:

$$ \begin{align*}
    & \left( \begin{array}{cc|c}
        1 & -1 & 1 \\
        1 & 1 & 3
    \end{array} \right)
    \begin{array}{l} \\ R_2 - R_1 \end{array}
    \longrightarrow  &
    \left( \begin{array}{cc|c}
        1 & -1 & 1 \\
        0 & 2 & 2
    \end{array} \right)
\end{align*} $$

Solving by back substitution gives $x_1 = 2$ and $x_2 = 1.$

```{prf:theorem} Consistent systems
:label: consistent-system-theorem

A system of linear equations $A\mathbf{x} = \mathbf{b}$ is consistent if $\operatorname{rank}(A) = \operatorname{rank}(A \mid \mathbf{b})$.
```

## Inconsistent systems

If the lines that represent each equation in a system with two variables never intersect, then they must be parallel - and we have an **inconsistent system**. Consider the following system of linear equations

$$ \begin{align*}
    x_1 - x_2 &= 1, \\
    x_1 - x_2 & = -1.
\end{align*} $$

The lines for these two equations have been plotted in {numref}`inconsistent-system-plot` below.

```{glue:figure} inconsistent_system_plot
:figwidth: 500
:name: inconsistent-system-plot

Plots of the lines $x_1 - x_2 = 1$ and $x_1 - x_2 = -1$. 
```

The two lines do not intersect, so this system does not have a solution. If we attempt to solve this using Gaussian elimination, we find:

$$ \begin{align*}
    & \left( \begin{array}{cc|c}
        1 & -1 & 1 \\
        1 & -1 & -1
    \end{array} \right)
    \begin{array}{l} \\ R_2 - R_1 \end{array} &
    \longrightarrow &
    \left( \begin{array}{cc|c}
        1 & -1 & 1 \\
        0 & 0 & -2
    \end{array} \right),
\end{align*} $$

Here, the second equation is $0x_1 + 0x_2 = -2$, which is clearly impossible.

```{prf:theorem} Inconsistent systems
:label: inconsistent-system-theorem

A system of linear equations is inconsistent if $A\mathbf{x} = \mathbf{b}$ is inconsistent if $\operatorname{rank}(A) < \operatorname{rank}(A \mid \mathbf{b})$.
```

## Indeterminate systems

If the lines that represent each equation in a system with two variables are the same line, than we have an **indeterminate system**, where we have an infinite number of solutions. Consider the following system of linear equations

$$ \begin{align*}
    v.
\end{align*} $$

The lines for these two equations have been plotted in {numref}`indeterminate-system-plot` below.

```{glue:figure} indeterminate_system_plot
:figwidth: 500
:name: indeterminate-system-plot

Plots of the lines $x_1 - x_2 = 1$ and $2x_1 - 2x_2 = 2$. 
```

The two lines are the same - they are defined by different equations, but the set of points satisfying each equation is the same, so any point along either line represents a solution to the system. If we attempt to solve this using Gaussian elimination, we get:

$$ \begin{align*}
    \left( \begin{array}{cc|c}
        1 & -1 & 1 \\
        2 & -2 & 2
    \end{array} \right)
    \begin{array}{l} \\ R_2 - 2R_1 \end{array} 
    \longrightarrow \quad
    \left( \begin{array}{cc|c}
        1 & -1 & 1 \\
        0 & 0 & 0 
    \end{array} \right),
\end{align*} $$

Here, the second equation has been eliminated, so we have a single equation $x_1 - x_2 = 1$. This is not enough to define a unique pair of points $x_1$ and $x_2$, so any set of points satisfying this one equation will be a valid solution to the system.

```{prf:theorem} indeterminate systems
:label: indeterminate-system-theorem

A system of linear equations with $n$ unknowns $A\mathbf{x} = \mathbf{b}$ is indeterminate if $\operatorname{rank}(A) < n$.
```

If we have an indeterminate system, the variables that correspond to the columns in the coefficient matrix without pivots (i.e., variables we cannot solve for) are called **free variables**. We express the solution of indeterminate systems by assigning a parameter ($r$, $s$, $t$ etc.) to the free variables, and proceed as normal. For example, in our system, $x_2$ is a free variable - so let $x_2 = r$ where $r \in \mathbb{R}$ and

$$ x_1 - r = 1, $$

so the solution is $x_1 = 1 + r$ and $x_2 = r$. Choosing any value of $r$ gives a solution to the system, representing a point on the line of possible solutions.

For example, if $r=1$, then $x_1 = 2$ and $x_2 = 1$. Alternatively if $r = -1$ then $x_1 = 0$ and $x_2 = -1$.

```{prf:example}
:label: rank-example

Determine the rank of the coefficient matrix and the augmented matrix for the following systems of linear equations and classify them as either consistent, inconsistent or indeterminate systems.
    
(i) &emsp; $\begin{array}{rl}
    3x_1 + x_2 - 2x_3  \!\!\!\! &= 1 \\
    x_1 - x_2 + 2x_3  \!\!\!\! &= 3 \\
    2x_1 - 3x_2 + 7x_3  \!\!\!\! &= 4
\end{array}$ &emsp;

(ii) &emsp; $\begin{array}{rl}
    x_1 - x_2 + 2x_3  \!\!\!\! &= 3 \\
    2x_1 - 3x_2 + 7x_3  \!\!\!\! &= 4 \\
    -x_1 + 3x_2 - 8x_3  \!\!\!\! &= 1
\end{array}$ &emsp;

(iii) &emsp; $\begin{array}{rl}
    x_1 + x_2 - 2x_3  \!\!\!\! &= 1 \\
    2x_1 - x_2 + x_3  \!\!\!\! &= 9 \\
    x_1 + 4x_2 - 7x_3  \!\!\!\! &= 2
\end{array}$ &emsp;

```{dropdown} Solution

(i) &emsp; 

$$ \begin{align*}
    & \left( \begin{array}{ccc|c}
        3 & 1 & -2 & 1 \\
        1 & -1 & 2 & 3 \\
        2 & -3 & 7 & 4
    \end{array} \right)
    \\ \\
    R_1 \leftrightarrow R_2 \longrightarrow &
    \left( \begin{array}{ccc|c}
        1 & -1 & 2 & 3 \\
        3 & 1 & -2 & 1 \\
        2 & -3 & 7 & 4
    \end{array} \right)
    \\ \\
    \begin{array}{l} R_2 - 3R_1 \\ R_3 - 2R_1 \end{array} \longrightarrow &
    \left( \begin{array}{ccc|c}
        1 & -1 & 2 & 3 \\
        0 & 4 & -8 & -8 \\
        0 & -1 & 3 & -2
    \end{array} \right)
    \\ \\
    R_3 + \frac{1}{4}R_2 \longrightarrow &
    \left( \begin{array}{ccc|c}
        1 & -1 & 2 & 3 \\
        0 & 4 & -8 & -8 \\
        0 & 0 & 1 & -4
    \end{array} \right)
\end{align*} $$

The augmented matrix is now in row echelon form. Since $\operatorname{rank}(A) = \operatorname{rank}(A \mid \mathbf{b}) = 3$, this is a consistent system by the {prf:ref}`theorem for consistent systems<consistent-system-theorem>`. Using back-substitution, the solution is $x_1 = 1$, $x_2 = -10$ and $x_3 = -4$. 

(ii)

$$ \begin{align*}
    & \left( \begin{array}{ccc|c}
        1 & -1 & 2 & 3 \\
        2 & -3 & 7 & 4 \\
        -1 & 3 & -8 & 1 
    \end{array} \right)
    \\ \\
    \begin{array}{l} R_2 - 2R_1 \\ R_3 + R_1 \end{array} \longrightarrow &
    \left( \begin{array}{ccc|c}
        1 & -1 & 2 & 3 \\
        0 & -1 & 3 & -2 \\
        0 & 2 & -6 & 4 
    \end{array} \right)
    \\ \\
     R_3 + 2R_2 \longrightarrow &
    \left( \begin{array}{ccc|c}
        1 & -1 & 2 & 3 \\
        0 & -1 & 3 & -2 \\
        0 & 0 & 0 & 0
    \end{array} \right)
\end{align*} $$

The augmented matrix is now in row echelon form. Since $\operatorname{rank}(A) = \operatorname{rank}(A \mid \mathbf{b}) = 2$, this is a consistent system by the {prf:ref}`theorem for consistent systems<consistent-system-theorem>` - that is, it has a solution. Furthermore, since $\operatorname{rank}(A) $ is less than the number of unknowns, this is an indeterminate system by the {prf:ref}`theorem for indeterminate systems<indeterminate-system-theorem>`, so it has infinitely many solutions. Let us define a parameter by $x_3 = r$, then using back-substitution we have $x_1 = 5 + r$ and $x_2 = 2 + 3r$.

(iii)

$$ \begin{align*}
    & \left( \begin{array}{ccc|c}
        1 & 1 & -2 & 1 \\
        2 & -1 & 1 & 9 \\
        1 & 4 & -7 & 2 
    \end{array} \right) 
    \\ \\
    \begin{array}{l} \\ R_2 - 2R_1 \\ R_3 - R_1 \end{array} \longrightarrow &
    \left( \begin{array}{ccc|c}
        1 & 1 & -2 & 1 \\
        0 & -3 & 5 & 7 \\
        0 & 3 & -5 & 1 
    \end{array} \right)
    \\ \\
    R_3 + R_2 \longrightarrow &
    \left( \begin{array}{ccc|c}
        1 & 1 & -2 & 1 \\
        0 & -3 & 5 & 7 \\
        0 & 0 & 0 & 8
    \end{array} \right)
\end{align*} $$

The augmented matrix is now in row echelon form, and we have $\operatorname{rank}(A) = 2$ and $\operatorname{rank}(A \mid \mathbf{b}) = 3$. Since $\operatorname{rank}(A) < \operatorname{rank}(A \mid \mathbf{b})$ then this is an inconsistent system by the {prf:ref}`theorem for inconsistent systems<inconsistent-system-theorem>`, and does not have a solution.
```

````python
import numpy as np
import matplotlib.pyplot as plt
from myst_nb import glue
# plt.rcParams['text.usetex'] = True

# Consistent system
x = np.linspace(0, 4, 200)
y2 = 3 - x
y1 = -1 + x

fig, ax = plt.subplots()
plt.plot(x, y1, 'b', label="$x_1 + x_2 = 3$")
plt.plot(x, y2, 'r', label="$x_1 - x_2 = 1$")
plt.plot(2, 1, "ko")
plt.axis([0, 4, 0, 4])
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.legend()
# plt.show()

plt.savefig("../_images/2_consistent_systems.png", bbox_inches='tight')
glue("consistent_system_plot", fig, display=False)

# Inconsistent system
y1 = x + 1
y2 = x - 1

fig, ax = plt.subplots()
plt.plot(x, y1, 'b', label="$x_1 - x_2 = -1$")
plt.plot(x, y2, 'r', label="$x_1 - x_2 = 1$")
plt.axis([0, 4, 0, 4])
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.legend()
plt.show()

glue("inconsistent_system_plot", fig, display=False)

# indeterminate system
y1 = x + 1
y2 = (2 * x + 2) / 2

fig, ax = plt.subplots()
plt.plot(x, y1, 'b', label="$x_1 - x_2 = 1$")
plt.plot(x, y2, 'r--', label="$2x_1 - 2x_2 = 2$")
plt.axis([0, 4, 0, 4])
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.legend()
plt.show()

glue("indeterminate_system_plot", fig, display=False)
````
